```python
import pandas
import matplotlib.pyplot as plt
data = pandas.read_csv("API_EN.ATM.CO2E.KT_DS2_en_csv_v2_5871652.csv",  delimiter=',' , skiprows = 4) # make sure that the location is same as your directory
df = data.copy()
```

```python
df
```

```python
df['Country Name'].unique()
```

```python
# input the value of as given above
while True : 
    country_name = input('Enter a country name: ')

    if country_name in df['Country Name'].unique():
        break

    print(f'Country "{country_name}" not found in the dataset.')
```

```python
# run this to plot the graph of above mention country
country_data = df[df['Country Name'] == country_name]
years = country_data.columns[4:-1] 
emissions = country_data.iloc[:, 4:-1].values.squeeze() 

plt.figure(figsize=(12, 6))
plt.plot(years, emissions, marker='o', linestyle='-')
plt.title(f'CO2 Emissions for {country_name}')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (kt)')
plt.grid(True)
plt.xticks(rotation= 90)
plt.tight_layout()
plt.show()
```
